# pendapatan (dalam Rupiah) berdasarkan pengalaman kerja (dalam tahun).

LIBRARY:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Contoh Dataset:
Pengalaman kerja dalam tahun, pendapatan dalam rupiah

```python
X = np.array([1, 2, 3, 4, 5, 6, 8, 10])       # pengalaman (tahun)
y = np.array([3, 5, 7, 9, 11, 13, 17, 21])    # gaji (juta Rupiah)

# Jumlah data
m = X.shape[0]
print(f"Jumlah data: {m}")
```

VISUALISASI DATA

```python
plt.scatter(X, y, color='blue', marker='x', label="Data Asli")
plt.xlabel("Pengalaman Kerja (tahun)")
plt.ylabel("Pendapatan (juta Rupiah)")
plt.title("Hubungan antara Pengalaman dan Pendapatan")
plt.legend()
plt.show()
```

FUNGSI MODEL LINIER

```python
def compute_model_output(X, w, b):
    """
    Menghitung prediksi model linear f_wb = w*x + b
    """
    return w * X + b
```

BIAYA/COST FUNCTION

```python
def compute_cost(X, y, w, b):
    """
    Menghitung Mean Squared Error (MSE)
    """
    m = X.shape[0]
    total_cost = 0

    for i in range(m):
        f_wb = w * X[i] + b
        total_cost += (f_wb - y[i]) ** 2

    return total_cost / (2 * m)
```

GRADIEN DESCENT

```python
def compute_gradient(X, y, w, b):
    """
    Menghitung turunan (gradien) dari fungsi biaya terhadap w dan b
    """
    m = X.shape[0]
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = w * X[i] + b
        dj_dw += (f_wb - y[i]) * X[i]
        dj_db += (f_wb - y[i])

    dj_dw /= m
    dj_db /= m

    return dj_dw, dj_db

def gradient_descent(X, y, w_in, b_in, alpha, num_iters, compute_cost, compute_gradient):
    """
    Menjalankan proses gradient descent untuk menemukan w dan b terbaik
    """
    J_history = []
    w = w_in
    b = b_in

    for i in range(num_iters):
        dj_dw, dj_db = compute_gradient(X, y, w, b)

        w = w - alpha * dj_dw
        b = b - alpha * dj_db

        if i < 10000:  # batasi simpan riwayat biaya
            J_history.append(compute_cost(X, y, w, b))

        if i % 100 == 0:
            print(f"Iterasi {i}: Cost = {J_history[-1]:.4f}, w = {w:.4f}, b = {b:.4f}")

    return w, b, J_history
```

INSIALISASI DAN TRAINING MODEL

```python
w_init = 0
b_init = 0
alpha = 0.01
iterations = 1000

w_final, b_final, J_hist = gradient_descent(X, y, w_init, b_init, alpha, iterations, compute_cost, compute_gradient)

print("\nHasil Akhir Training:")
print(f"w = {w_final:.4f}")
print(f"b = {b_final:.4f}")
```

VISUALISASI HASIL

```python
predictions = compute_model_output(X, w_final, b_final)
plt.plot(X, predictions, color='red', label='Model Prediksi')
plt.scatter(X, y, color='blue', marker='x', label='Data Asli')
plt.xlabel("Pengalaman Kerja (tahun)")
plt.ylabel("Pendapatan (juta Rupiah)")
plt.title("Regresi Linier: Prediksi Pendapatan vs Pengalaman")
plt.legend()
plt.show()
```

PREDIKSI BARU

```python
pengalaman_baru = 7
gaji_prediksi = compute_model_output(pengalaman_baru, w_final, b_final)
print(f"Perkiraan pendapatan untuk pengalaman {pengalaman_baru} tahun = Rp {gaji_prediksi*1_000_000:.0f}")
```
